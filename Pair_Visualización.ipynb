```python
# importamos las librerías que necesitamos

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd

# Visualización
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
# lo primero que hacemos es cargar el dataframe que vamos a usar para esta lección
df = pd.read_csv("world_data_full_apply_nulos.csv") 
df.head()
```

```python
df.columns
```

```python
# ¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?

sns.histplot(x = "birth_rate", data = df, color = "violet")
```

```python
# ¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida? Proporciona una breve descripción de los patrones que observes.

# Podemos observar que la concentración de los datos está en la variable "life_expectative", por lo que deducimos que hay más esperanza de vida, 
# que tasa de mortalidad infaltil, esto asumiendo que el "life_expectative" es años y el "infant_mortality" es porcentaje. 

sns.scatterplot(x = "infant_mortality", 
                y = "life_expectancy", 
                data = df)
```

```python
# ¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente.

sns.countplot(x = "continent", 
              data = df, 
              color = "c")
```

```python
df_conteo = df["continent"].value_counts().reset_index()

df_conteo
```

```python
plt.pie("count", labels="continent", data=df_conteo, autopct=  '%1.2f%%')
```

```python
# ¿Cuál es la mediana de la esperanza de vida en Europa? Proporciona un rango intercuartil aproximado y verifica si existen valores atípicos.

sns.boxplot(y= "continent", x= "life_expectancy", data = df)
```
